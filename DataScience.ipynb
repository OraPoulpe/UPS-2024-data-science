```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
cs = pd.read_csv(filepath_or_buffer = 'marketing_campaign.csv', delimiter='\t')
cs
```

### Построение графиков

```python
# 1 График

sns.relplot(data=cs, x="Year_Birth", y="Income")
```

```python
# 2 График

plt.figure(figsize=(25, 5))
sns.lineplot(data=cs, x="Year_Birth", y="Income")
```

```python
# 3 График

plt.figure(figsize=(25, 5))
sns.lineplot(data=cs, x="Marital_Status", y="Income")
```

```python
# 4 График

sns.countplot(data=cs, x="Marital_Status", hue="Kidhome")
```

```python
# График 5

plt.figure(figsize=(60,8))
sns.lineplot(data=cs, x="MntGoldProds", y="Income")
```

```python
# График 6

plt.figure(figsize=(30,10))
sns.barplot(data=cs, x="Year_Birth", y="NumWebPurchases")
```

```python
# График 7

plt.figure(figsize=(35,10))
sns.barplot(data=cs, x="Year_Birth", y="NumStorePurchases")
```

### Корреляция

```python
matrix = cs.corr(numeric_only=True).round(2)
plt.figure(figsize=(18,18))
sns.heatmap(matrix, annot = True, vmin=-1, vmax=1, center= 0)
plt.show()
```

Вывод
1) Есть зависимость между доходами и тратами на продукты
2) Есть зависимость между покупками различных товаров
